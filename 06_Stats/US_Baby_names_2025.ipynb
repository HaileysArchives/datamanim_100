## Group the dataset by name and assign to names.
> 데이터 집합을 이름으로 지정하고 이름에 할당

```python
df.info()
```

- value_counts()
- sort_values()

```python
names = df.drop(columns=['Year'], axis=1)

names = names.groupby('Name').sum()

# names.sort_values(by=['Count'], ascending=False)
names['Count'].sort_values(ascending=False)
```

## How many different names exist in the dataset?

```python
df['Name'].nunique()
```

```python
# as we have already grouped by the name, all the names are unique already.

len(names)
```

```python
df['Name'].value_counts()
```

## What is the name with most occurrences?

```python
names['Count'].idxmax()
```

## How many different names have the least occurences?

## What is the median name occurence?

```python
names.loc[names['Count'] == names['Count'].median(), :]
```
