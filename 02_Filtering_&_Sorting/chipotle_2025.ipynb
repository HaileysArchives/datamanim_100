```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
df.head(5)
```

## How many products cost more than $10,00?

```python
df.info()
```

```python
prices = df['item_price'].str.replace('$', '').astype('float')
prices.head(5)

df['item_price'] = prices
```

```python
df.info()
```

### 중복된 행 제거: drop_duplicates()

```python
df_filtered = df.drop_duplicates(['item_name', 'choice_description', 'quantity'])
df_filtered
```

```python
one_prod = df.loc[df['quantity'] == 1]
one_prod.head()
```

```python
ten_dollars = one_prod[one_prod['item_price'] > 10]['item_name']
ten_dollars
```

```python
ten_dollars.unique()
```

```python
ten_dollars.nunique()
```

## What is the price of each item?

```python
item_per_price = df.loc[df['quantity'] == 1, :]
item_per_price = item_per_price.drop(columns=['choice_description', 'quantity', 'order_id'])
per_price = item_per_price.sort_values(by='item_price', ascending=False)
per_price
```

- .drop(): pandas에서 행이나 열을 제거하는 데 사용되는 함수

```python
per_price['item_name'].unique()
```

```python
per_price.drop_duplicates(['item_name'])
```
