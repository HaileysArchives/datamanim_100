## Oops, it seems our first dataset has some unnamed blank collumns, fix cars1.
> 첫 번째 데이터 집합에 이름이 지정되지 않는 빈 열이 있는 것 같아. cars1 데이터를 수정해줘.

```python
cars1.isnull().sum()
```

```python
cars1 = cars1.loc[:, 'mpg':'car']
cars1.head()
```

## What is the number of observations in each dataset?

## Join cars1 and cars2 into a single DataFrame called cars.

- 열 이름이 일치하지 않는 경우 해당 열은 포함되지 않거나 새 열로 추가된다.
- pd.merge( ): 공통 열 기준으로 병합
> - merge는 항상 '행(row)'을 기준으로 결합
- pd.concat( ): 단순 연결 → 열 개수 달라도 자동으로 NaN 채움.

```python
cars = pd.merge(cars1, cars2, how='outer')
```

```python
cars
```

```python
# cars = cars1.append(cars2)
```

- append( ): 데이터프레임을 **'행 방향'**으로 단순히 추가하며, 열 이름이 달라도 NaN으로 채움.

- pd.merge( ): **'공통 열(키)'**을 기준으로 데이터프레임을 병합하며, 다양한 조인 방식(inner, outer 등)을 지원.

- pd.concat( ): 여러 데이터프레임을 **'축(axis) 기준'**으로 단순히 연결하며, 열 이름이 달라도 NaN으로 채움.
