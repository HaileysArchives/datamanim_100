```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.Series(np.random.randint(1, high=5, size = 100, dtype='l'))
df2 = pd.Series(np.random.randint(1, high=4, size=100, dtype='l'))
df3 = pd.Series(np.random.randint(10000, high=30001, size=100, dtype='l'))

df1
```

Let's create a DataFrame by joining the Series by column
> 열을 기준으로 Series를 조인하여 데이터 프레임을 만들어라.

```python
house = pd.concat([df1, df2, df3], axis = 1)
house.head()
```

```python
house.rename(columns={0:'bedrs', 1:'bathrs', 2:'price_sqr_meter'})
house.head()
```

Change the name of the columns to berds, bathrs, price_sqr_meter

```python
house.rename(columns={0:'bedrs', 1:'bathrs', 2:'price_sqr_meter'}, inplace=True)
# inplace=True 해당 메서드가 원본 데이터프레임을 직접 수정
# inplace=False 새로운 데이터프레임을 반환받고 이를 다시 house에 할당
house.head()
```

Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
> 3 시리즈의 값으로 하나의 열 데이터 프레임을 만들고 'bigcolumn'에 할당합니다.

```python
bigcolumn = pd.concat([df1, df2, df3], axis=0)

# pd.concat()를 사용해서 여러 Series를 연결해도 그 결과는 여전히 Series 객체이다.
# 왜냐하면 concat() 메소드는 Series를 축을 기준으로 결합할 때 기본적으로 결과가 Series 객체로 반환

# type(bigcolumn)

bigcolumn = bigcolumn.to_frame()
print(type(bigcolumn))

bigcolumn
```

- to_frame(): Pandas Series => DataFrame으로 변환
- to_series(): Pandas DataFrame의 특정 칼럼을 Series로 변환
- to_dict(): Pandas DataFrame을 딕셔너리로 변환

it seems it is going only until index 99. Is it true?

```python
len(bigcolumn)
```

Reindex the DataFrame so it goes from 0 to 299

```python
bigcolumn = bigcolumn.reset_index(drop=True, inplace=True)
bigcolumn

# inplace=True로 설정하면 reset_index() 메소드는 새로운 DataFrame을 반환하지 않고 None을 반환
```

```python
bigcolumn.head()
```

```python
bigcolumn = bigcolumn.reset_index(drop=True) # drop=True 기존 인덱스를 열로 추가하지 않고 삭제
bigcolumn
```

- pd.concat()은 여러 DataFrame이나 Series를 결합하는 함수
> 각 DataFrame 또는 Series의 행을 이어붙이는 방식
- reset_index()는 인덱스를 초기화하고 기존의 인덱스를 열로 추가하는 메서드
> drop=True를 설정하면 기존 인덱스를 열로 추가하지 않고 삭제

**왜 bigcolumn이 100개에서 300개로 늘어났는가?**
> reset_index(drop=True)는 단지 기존 인덱스를 삭제하고, 새로운 인덱스를 할당하는 역할을 하며, 데이터의 행 수에는 영향을 미치지 않습니다.
