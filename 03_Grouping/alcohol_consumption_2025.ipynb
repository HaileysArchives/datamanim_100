## Which continent drinks more beer on average?

- groupby() 메소드: 그룹화된 데이터를 다룰 수 있는 인터페이스를 제공
- 따라서 head(n)로 반환된 데이터프레임은 그룹별로 n개의 행을 포함할 수 있기 때문에, 전체적으로는 n보다 더 많은 행이 출력될 수 있음

```python
beer_avg = df.groupby('continent')
beer_avg['beer_servings'].mean()
```

## For each continent print the statistics for wine consumption.
