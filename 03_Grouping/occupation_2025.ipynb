```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
df.head()
```

```python
df = df.set_index(keys='user_id')
df.head()
```

## Discover what is the mean age per occupation.

```python
mean_age = df.groupby('occupation')['age'].mean()
mean_age
```

## Discover the Male ratio per occupation and sort it from the most to the least.

```python
df.info()
```

```python
male = df.loc[df['gender'] == 'M']
male = male.groupby('occupation')
male.head()
```

### the Male ratio per occupation 주의하기!

```python
# 먼저 male만 덜어내기!

def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

df['gender_n'] = df['gender'].apply(gender_to_numeric)

occup_ratio = df.groupby('occupation')['gender_n'].sum() / df['occupation'].value_counts() * 100
occup_ratio.sort_values(ascending=False)
```

## For each occupation, calculate the minimum and maximum ages.

- agg(): aggregation의 줄임말로, 데이터를 요약하거나 집계하는 과정을 말한다.
- apply(function)과 헷갈리지 말기!

```python
df.groupby('occupation')['age'].agg(['min', 'max'])
```

## For each combination of occupation and gender, calculate the mean age.

```python
df.groupby(['occupation', 'gender'])['age'].mean(numeric_only=True)
```

- 숫자형 데이터에만 mean()을 적용

```python
df.groupby(['occupation', 'gender'])['age'].agg(['mean'])
```

## For each occupation present the percentage of women and men.
> 각 직종별로 여성과 남성의 비율을 표시

```python
df['gender_encoded'] = df['gender'].map({'M':1, 'F':0})

male_count = df[df['gender_encoded'] == 1].groupby('occupation')['gender_encoded'].sum()
female_count = df[df['gender_encoded'] == 0].groupby('occupation')['gender_encoded'].sum()

total_count = df.groupby('occupation').size()

male_ratio = (male_count / total_count) * 100
female_ratio = (female_count / total_count) * 100

result = pd.DataFrame({'occupation': male_ratio.index, 'male' : male_ratio.values, 'female' : female_ratio.values})
result
```

```python
gender_ocup = df.groupby(['occupation', 'gender']).agg({'gender': 'count'})
gender_ocup
```

```python
occup_count = df.groupby(['occupation']).agg('count')

occup_gender = gender_ocup.div(occup_count, level='occupation') * 100
occup_gender.loc[: , 'gender']
```

- map() 함수 사용: 특정 값에 대해 1:1 매핑을 제공하여 데이터를 변환하는 데 적합
> df['gender_encoded'] = df['gender'].map({'M':1, 'F':0})

- get_dummies(): 범주형 함수를 자동으로 더미 변수 열로 반환. df.get_dummies(df, columns=['gender'])
- replace(): df['gender'].replace({'M':1, 'F':0})
