```python
import pandas as pd
import numpy as np
```

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
df = pd.DataFrame(raw_data, columns=raw_data.keys())
df.head()
```

```python
df.info()
```

## Assign it to a variable called regiment.

## What is the mean preTestScore from the regiment Nighthawks?

```python
df[df['regiment'] == 'Nighthawks'].groupby('regiment').mean(numeric_only=True)
```

```python
nighthawks = df.loc[df['regiment'] == 'Nighthawks', 'preTestScore']
nighthawks.mean()
```

## Present general statistics by company.

```python
company_stat = df.groupby('company').describe()
company_stat
```

## What is the mean of each company's preTestScore?

```python
each_com = df.groupby('company')['preTestScore'].mean()
each_com
```

## Present the mean preTestScores grouped by regiment and company.


```python
df.groupby(['regiment', 'company'])['preTestScore'].mean()
```

## Group the entire dataframe by regiment and company.

```python
entire = df.groupby(['regiment', 'company']).mean(numeric_only=True)
entire
```

## What is the number of observations in each regiment and company.

```python
df.groupby(['regiment', 'company']).size()
```

- size(): DataFrame에서 전체 요소 개수(값의 개수)를 반환하는 메소드

## Iterate over a group and print the name and the whole data from the regiment.
> 그룹을 반복하고 연대의 이름과 전체 데이터를 출력

```python
copy = df.groupby('regiment')
copy.head()
```

```python
# Iterate over a group
for name, group in copy:
    print(name)
    print(group)
```

- name: 그룹화에 사용된 regiment 열의 고유값: 각 그룹을 대표하는 이름
- group: 해당 그룹에 속하는 DataFrame 부분
