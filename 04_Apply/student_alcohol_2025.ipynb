## Create a lambda function that will capitalize strings.

```python
capitalizer = lambda x: x.capitalize()
```

- upper() vs capitalize()
- upper() method: 문자열의 모든 글자를 대문자로 변환
- capitalize() method: 첫 글자만 대문자로, 나머지는 소문자로 바꾸는 데 초점

## Capitalize both Mjob and Fjob.

```python
df['Mjob'] = df['Mjob'].apply(capitalizer)
df['Fjob'] = df['Fjob'].apply(capitalizer)
```

```python
df.loc[:, 'Mjob':'Fjob']
```

## Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old.)
> legal_drinker라는 새 열에 부울 값을 반환하는 majority라는 함수를 만듭니다(과반수를 17세 이상으로 간주).

```python
def majority(x):
    if x >= 17:
        return True
    else:
        return False
```

```python
df['legal_drinker'] = df['age'].apply(majority)
df.loc[:, 'legal_drinker']
```

## Multiply every number of the dataset by 10.

```python
def multi(x):
    if type(x) is int:
        return 10 * x
    return x
```

```python
df.apply(multi)
```

> 함수 multi는 개별 값에 대해 동작하도록 설계되었지만, apply()는 한 번에 각 열(또는 행)을 통째로 함수에 전달하려고 해서 제대로 작동하지 않음.

## apply() vs applymap()
- apply(): Series 객체에도 사용 및 DataFrame에서 열 또는 행 단위로 함수 적용
- applymap(): DataFrame의 각 요소(elements)에 대해 함수 적용 (DataFrame 전용 메소드)

```python
df.applymap(multi)
```
